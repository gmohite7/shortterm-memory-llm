```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
```

```python
namespace = ('user','u1')
```

```python
# creating memories
store.put(namespace,'1',{'data':'Users name is Girish'})
store.put(namespace,'2',{'data':'User likes AI Engg'})
```

```python
store.put(namespace,'2',{'data':'User likes biryani'})
```

```python
#another namespace
namespace2 = ('user','u2')

store.put(namespace2,'1',{'data':'user likes pasta'})
store.put(namespace2,'1',{'data':'user prefers grid style navigation'})
```

```python
# retrieving memory

memories = store.get(namespace,'2')

print(memories)
```

```python
# retrieving all memories

all_memories = store.search(namespace)

for i in all_memories:
    print(i)
```
